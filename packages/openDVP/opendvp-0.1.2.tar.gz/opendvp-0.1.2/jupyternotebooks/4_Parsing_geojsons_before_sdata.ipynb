```python
#imports
import pandas as pd
import geopandas as gpd
import os
import time
import ast
```

# Scenario 1

```python
#paths
path_to_geojson_1 = "../test_data/geojsons/20241010_1340_991.geojson"
```

```python
gdf = gpd.read_file(path_to_geojson_1)
gdf.head()
```

```python
# Classification describes the QuPath class of the various annotations objects.
# We have to remove the Point annotations (seen as POINT in the geometry column)
    # print how many points
print(f"Found {len(gdf[gdf.geom_type == 'Point'])} points in geojson, these are now removed")
gdf = gdf[gdf.geom_type != 'Point']
gdf.reset_index(drop=True, inplace=True)
```

```python
# We have to extract the class name from the classification column
    #check if classification column is a string or dictionary
if isinstance(gdf['classification'][0], (str)):
    gdf['class'] = gdf['classification'].apply(lambda x: ast.literal_eval(x)['name'])
elif isinstance(gdf['classification'][0], (dict)):
    gdf['class'] = gdf['classification'].apply(lambda x: x['name'])
```

```python
gdf.head()
```

```python
gdf['class'].value_counts()
```

# Scenario 2

```python
path_to_geojson_1 = "../data/geojsons/20241010_1340_991.geojson"
path_to_geojson_2 = "../data/geojsons/20241011_1031_991_refill.geojson"
```

```python
gdf_1 = gpd.read_file(path_to_geojson_1)
gdf_2 = gpd.read_file(path_to_geojson_2)
gdf_1.head()
```

```python
gdf_2.head()
```

```python
def remove_points_from_gdf(gdf):
    print(f"Found {len(gdf[gdf.geom_type == 'Point'])} points in geojson, these are now removed")
    gdf = gdf[gdf.geom_type != 'Point']
    gdf.reset_index(drop=True, inplace=True)
    return gdf
```

```python
gdf_1 = remove_points_from_gdf(gdf_1)
gdf_2 = remove_points_from_gdf(gdf_2)
```

```python
gdf_1.head()
```

```python
gdf_2.head()
```

```python
def extract_class_from_classification(gdf):
    if isinstance(gdf['classification'][0], (str)):
        gdf['class'] = gdf['classification'].apply(lambda x: ast.literal_eval(x)['name'])
    elif isinstance(gdf['classification'][0], (dict)):
        gdf['class'] = gdf['classification'].apply(lambda x: x['name'])
    return gdf

gdf_1 = extract_class_from_classification(gdf_1)
gdf_2 = extract_class_from_classification(gdf_2)
```

```python
gdf_1.head()
```

```python
gdf_2.head()
```

```python
gdf_1['class'].value_counts()
```

```python
gdf_2['class'].value_counts()
```

```python
# these are non overlapping classes, so we can merge the two dataframes
gdf = pd.concat([gdf_1, gdf_2], ignore_index=True)
```

```python
gdf['class'].value_counts()
```

# Scenario 3

```python
path_to_geojson_1 = "../data/geojsons/20241010_1340_991.geojson"
path_to_geojson_2 = "../data/geojsons/20241011_1031_991_refill.geojson"

gdf_1 = gpd.read_file(path_to_geojson_1)
gdf_2 = gpd.read_file(path_to_geojson_2)

gdf_1 = remove_points_from_gdf(gdf_1)
gdf_2 = remove_points_from_gdf(gdf_2)

gdf_1 = extract_class_from_classification(gdf_1)
gdf_2 = extract_class_from_classification(gdf_2)
```

```python
# these time the classes are going to the same wells
# so we have to concatenate the two dataframes
# then create a new column with harmonized class names with a mapping dictionary
```

```python
gdf = pd.concat([gdf_1, gdf_2], ignore_index=True)
```

```python
gdf.head()
```

```python
#list all unique classes, export as csv for editing in Excel
df = pd.DataFrame(data=gdf['class'].unique().tolist())
df.to_csv("../data/geojsons/unique_classes.csv", index=False, header=False)
```

```python
df2 = pd.read_csv("../data/geojsons/unique_classes_mod.csv", sep=";") #the separator depends on software used to edit the csv
```

```python
df2
```

```python
category_mapping = df2.set_index('key')['value'].to_dict()
```

```python
category_mapping
```

```python
#use the mapping dictionary to create a new column with the harmonized class names
gdf['harmonized_class'] = gdf['class'].map(category_mapping)
```

```python
gdf.head()
```

```python
#export new geojson with harmonized class names
datetime = time.strftime("%Y%m%d_%H%M")
gdf.to_file(f"../data/geojsons/{datetime}_Primary991_harmonized.geojson", driver='GeoJSON')
```

# Caveats

We can only do so much from geojsons, whatever key we use (class or harmonized_class) has to be present in the metadata file of the proteomic dataset
